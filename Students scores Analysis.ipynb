### gender countchart

```python
pl=sns.countplot(data=df, x="Gender", color='pink')
for bar in pl.containers:
    pl.bar_label(bar)
```

```python
# from the above chart we have analysed that:
# the number of females in the data is more than the number of males
```

```python
sc=df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":"mean","WritingScore":'mean'})
sc
```

```python
sns.heatmap(sc, annot=True)
plt.title("Parent's Education and Student's Scores")
```

```python
# from the avobe chart we have concluded that the education of the parents
#have a good impact of their scores
```

```python
pm=df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":"mean","WritingScore":'mean'})
pm
```

```python
sns.heatmap(pm, annot=True)
plt.show()
```

```python
# from the avobe chart we have concluded that their is no/negligible
#impact of student's scores due to their parent's marital status
```

### Ethnic Group

```python
pl=sns.countplot(data=df,x="EthnicGroup")
for bar in pl.containers:
    pl.bar_label(bar)
```

```python
groupA=df.loc[(df['EthnicGroup']=="group A")].count()
groupB=df.loc[(df['EthnicGroup']=="group B")].count()
groupC=df.loc[(df['EthnicGroup']=="group C")].count()
groupD=df.loc[(df['EthnicGroup']=="group D")].count()
groupE=df.loc[(df['EthnicGroup']=="group E")].count()

mlist=[groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"], groupD["EthnicGroup"], groupE["EthnicGroup"]]
print(mlist)
l=["groupA", "groupB", "groupC", "groupD", "groupE" ]    

plt.pie(mlist,labels=l,autopct="%1.2f%%")
plt.title("Distribution of Ethnic Groups")
```
